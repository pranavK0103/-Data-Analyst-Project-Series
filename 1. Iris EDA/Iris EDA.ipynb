```python
# Calculate mean values for each numerical feature grouped by Species
mean_values = iris.groupby('Species').mean()
```

```python
mean_values
```

```python
plt.figure(figsize=(8,8))
plt.subplot(2, 2, 1)
plt.bar(list(mean_values["SepalLengthCm"].keys()),list(mean_values["SepalLengthCm"]))
plt.title('Mean Petal Width for Each Species')

plt.subplot(2, 2, 2)
plt.bar(list(mean_values["SepalWidthCm"].keys()),list(mean_values["SepalWidthCm"]))
plt.title('Mean Petal Width for Each Species')

plt.subplot(2, 2, 3)
plt.bar(list(mean_values["PetalLengthCm"].keys()),list(mean_values["PetalLengthCm"]))
plt.title('Mean Petal Width for Each Species')

plt.subplot(2, 2, 4)
plt.bar(list(mean_values["PetalWidthCm"].keys()),list(mean_values["PetalWidthCm"]))
plt.title('Mean Petal Width for Each Species')
plt.tight_layout()
plt.show()
```

```python
## kernal Plots
```

```python
# Set the style for the plots
sns.set(style="whitegrid")

# Plot kernel density plots for each numerical feature grouped by Species
plt.figure(figsize=(12, 8))

# Sepal Length
plt.subplot(2, 2, 1)
sns.kdeplot(data=iris, x='SepalLengthCm', hue='Species', fill=True, common_norm=False)
plt.title('Kernel Density Plot: Sepal Length by Species')

# Sepal Width
plt.subplot(2, 2, 2)
sns.kdeplot(data=iris, x='SepalWidthCm', hue='Species', fill=True, common_norm=False)
plt.title('Kernel Density Plot: Sepal Width by Species')

# Petal Length
plt.subplot(2, 2, 3)
sns.kdeplot(data=iris, x='PetalLengthCm', hue='Species', fill=True, common_norm=False)
plt.title('Kernel Density Plot: Petal Length by Species')

# Petal Width
plt.subplot(2, 2, 4)
sns.kdeplot(data=iris, x='PetalWidthCm', hue='Species', fill=True, common_norm=False)
plt.title('Kernel Density Plot: Petal Width by Species')

plt.tight_layout()
plt.show()
```

```python
#Stacked histogram
```

```python
# Set the style for the plots
sns.set(style="whitegrid")

# Plot stacked histograms for each numerical feature grouped by Species
plt.figure(figsize=(12, 8))

# Sepal Length
plt.subplot(2, 2, 1)
sns.histplot(data=iris, x='SepalLengthCm', hue='Species', multiple='stack', kde=True)
plt.title('Stacked Histogram: Sepal Length by Species')

# Sepal Width
plt.subplot(2, 2, 2)
sns.histplot(data=iris, x='SepalWidthCm', hue='Species', multiple='stack', kde=True)
plt.title('Stacked Histogram: Sepal Width by Species')

# Petal Length
plt.subplot(2, 2, 3)
sns.histplot(data=iris, x='PetalLengthCm', hue='Species', multiple='stack', kde=True)
plt.title('Stacked Histogram: Petal Length by Species')

# Petal Width
plt.subplot(2, 2, 4)
sns.histplot(data=iris, x='PetalWidthCm', hue='Species', multiple='stack', kde=True)
plt.title('Stacked Histogram: Petal Width by Species')

plt.tight_layout()
plt.show()
```

```python
## Swarm plot
```

```python
# Set the style for the plot
sns.set(style="whitegrid")

# Plot swarm plots for each numerical feature grouped by Species
plt.figure(figsize=(12, 8))

# Sepal Length
plt.subplot(2, 2, 1)
sns.swarmplot(x='Species', y='SepalLengthCm', data=iris, palette='Set1')
plt.title('Swarm Plot: Sepal Length by Species')

# Sepal Width
plt.subplot(2, 2, 2)
sns.swarmplot(x='Species', y='SepalWidthCm', data=iris, palette='Set1')
plt.title('Swarm Plot: Sepal Width by Species')

# Petal Length
plt.subplot(2, 2, 3)
sns.swarmplot(x='Species', y='PetalLengthCm', data=iris, palette='Set1')
plt.title('Swarm Plot: Petal Length by Species')

# Petal Width
plt.subplot(2, 2, 4)
sns.swarmplot(x='Species', y='PetalWidthCm', data=iris, palette='Set1')
plt.title('Swarm Plot: Petal Width by Species')

plt.tight_layout()
plt.show()
```
